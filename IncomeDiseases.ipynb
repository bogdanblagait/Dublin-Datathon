```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
#from matplotlib.style import style

plt.style.use('ggplot')

community_health = pd.read_csv("community_health.csv")

income = community_health.loc[community_health['indicator'] == "Median household income in US dollars"]
obesity = community_health.loc[community_health["indicator"] == "Age-adjusted percentage of adults overweight or obese (BMI 25 or higher)"]
diabetes = community_health.loc[community_health["indicator"] == "Age-adjusted percentage of adults with physician diagnosed diabetes"]
stroke = community_health.loc[community_health["indicator"] == "Age-adjusted percentage of adults with physician diagnosed angina, heart attack or stroke"]

for stat in [income,obesity,diabetes,stroke]:
    stat = stat[['county_name','precent_or_rate']]

income_obesity = pd.merge(income,obesity,on='county_name')
income_diabetes = pd.merge(income,diabetes,on='county_name')
income_stroke = pd.merge(income,stroke,on='county_name')

print(pearsonr(income_obesity['precent_or_rate_x'],income_obesity['precent_or_rate_y']))
#plt.scatter(income_obesity['precent_or_rate_x'],income_obesity['precent_or_rate_y'])

print(pearsonr(income_diabetes['precent_or_rate_x'],income_diabetes['precent_or_rate_y']))
#plt.scatter(income_diabetes['precent_or_rate_x'],income_diabetes['precent_or_rate_y'])

print(pearsonr(income_stroke['precent_or_rate_x'],income_stroke['precent_or_rate_y']))
plt.scatter(income_stroke['precent_or_rate_x'],income_stroke['precent_or_rate_y'])

plt.title("Income vs. Stroke/Heart Attack Rate")
plt.xlabel("Median Income")
plt.ylabel("Stroke/Heart Attack Rate")

plt.show()
```
